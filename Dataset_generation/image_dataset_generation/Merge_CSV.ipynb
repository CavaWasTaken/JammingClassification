```python
import os
import pandas as pd
```

```python
__basedir__ = os.path.abspath("")
path_of_file = os.path.join(__basedir__, 'image_dataset')
```

```python
csv_files = [ f for f in os.listdir(path_of_file) if f.endswith('.csv') ]
df_full = pd.DataFrame()

for file in csv_files:
    print(file)
    file = os.path.join(path_of_file, file)
    if os.path.isfile(file):
        print(file)
        df_ = pd.read_csv(file, sep=',', index_col=0)
        df_full = pd.concat([df_full, df_], ignore_index=True)
```

```python
df_full['path_spec'] = df_full['path_spec'].str.split('/home/arman-ebrahimi/Downloads/ATLANTIS_Dataset').str[1]
df_full
```

```python
unique_values = df_full['power'].unique()
unique_values
```

```python
df_full['power'] = df_full['power'].replace('0dB', 0)
df_full['power'] = df_full['power'].astype(int)
df_full
```

```python
CSN_final_name = os.path.join(__basedir__, 'features_all_.csv')
df_full.to_csv(CSN_final_name, sep=',')
```
