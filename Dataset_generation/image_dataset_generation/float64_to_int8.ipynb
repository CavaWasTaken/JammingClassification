**Convert the dataset from float64 to int8 in Python using NumPy**

```python
import os
import numpy as np
```

```python
def convert_float64_to_int8(matrix_npy):
    matrix_int8 = ((matrix_npy - matrix_npy.min()) / (matrix_npy.max() - matrix_npy.min()) * 255 - 128).astype(np.int8)
    return matrix_int8
```

```python
dataset_path = './array_dataset/'

# iterate over each folder in the dataset directory, representing different signal types
for signal_folder in os.listdir(dataset_path):
    print(f'Processing signal type folder: {signal_folder}')
    signal_folder_path = os.path.join(dataset_path, signal_folder)  # path to the signal type folder
    if os.path.isdir(signal_folder_path):
        # if the folder is not 'CLEAN', go one level deeper since it contains power subfolders
        if signal_folder != 'CLEAN':
            for power_folder in os.listdir(signal_folder_path):
                print(f'  Processing power subfolder: {power_folder}')
                power_folder_path = os.path.join(signal_folder_path, power_folder)  # path to the power subfolder
                if os.path.isdir(power_folder_path):
                    # for each file inside the power subfolder
                    for file in os.listdir(power_folder_path):
                        if file.endswith('.npy'):
                            file_path = os.path.join(power_folder_path, file)
                            # load the float64 matrix
                            matrix_float64 = np.load(file_path)
                            # convert to int8
                            matrix_int8 = convert_float64_to_int8(matrix_float64)
                            # save the int8 matrix
                            save_folder  = f'./image_dataset/{signal_folder}/{power_folder}'
                            os.makedirs(save_folder, exist_ok=True)
                            save_path = os.path.join(save_folder, file)
                            np.save(save_path, matrix_int8)

        # if the folder is 'CLEAN', process files directly inside it
        else:
            # for each file inside the subfolder
            for file in os.listdir(signal_folder_path):
                if file.endswith('.npy'):
                    file_path = os.path.join(signal_folder_path, file)
                    # load the float64 matrix
                    matrix_float64 = np.load(file_path)
                    # convert to int8
                    matrix_int8 = convert_float64_to_int8(matrix_float64)
                    # save the int8 matrix
                    save_folder  = f'./image_dataset/{signal_folder}'
                    os.makedirs(save_folder, exist_ok=True)
                    save_path = os.path.join(save_folder, file)
                    np.save(save_path, matrix_int8)
```
